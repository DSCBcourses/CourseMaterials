```python
# single imports cell
import numpy as np, imageio, os
from skimage.draw import ellipse
from skimage.filters import gaussian
from skimage.util import random_noise

os.makedirs(r'../Images/Napari', exist_ok=True)
OUT = r'../Images/Napari'
```

```python
# create three sample images
rng = np.random.default_rng(123)
for k in range(1,4):
    h,w = 512,512
    img = np.zeros((h,w), dtype=float)
    for i in range(45):
        cy = rng.integers(30, h-30); cx = rng.integers(30, w-30)
        r1 = rng.integers(6, 18); r2 = int(r1 * rng.uniform(0.6, 1.0))
        angle = rng.uniform(0, 2*np.pi)
        rr, cc = ellipse(cy, cx, r1, r2, rotation=angle, shape=img.shape)
        img[rr, cc] += rng.uniform(0.4, 1.0)
    if k==2:
        for cx in range(120,200,20):
            rr, cc = ellipse(200, cx, 18, 12, rotation=0, shape=img.shape)
            img[rr, cc] += 0.9
    img = gaussian(img, sigma=1.0)
    img = random_noise(img, mode='poisson')
    img = random_noise(img, mode='gaussian', var=0.003)
    img = (img - img.min())/(img.max()-img.min())
    img8 = (img*255).astype('uint8')
    path = os.path.join(OUT, f'sample{k}.png')
    imageio.imwrite(path, img8)
    print('wrote', path)
```
